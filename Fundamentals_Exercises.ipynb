## The questions:


* Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().

* One-hot encode the tensor you created in 9.








```python
import tensorflow as tf
```

```python
#Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().

scalar=tf.constant(7)
vector=tf.constant([4,3])
Matrix=tf.constant([[4,3],
                   [5,6],
                   [7,8]])
```

```python
#Find the shape, rank and size of the tensors you created in 1.
print(f"Rank of scalar = {scalar.ndim},shape={scalar.shape}")
print(f"Rank of vector = {vector.ndim},shape={vector.shape}")
print(f"Rank of Matrix = {Matrix.ndim},shape={Matrix.shape}")
```

```python
#Create two tensors containing random values between 0 and 1 with shape [5, 300].
#Using uniform to have random values spread over a range
tensor_A=tf.random.uniform([5,300])
tensor_B=tf.random.uniform([5,300])
tensor_A
tensor_B
```

```python
#Multiply the two tensors you created in 3 using matrix multiplication.
#we will be transposing the tensors instead of reshaping it.
mul=tf.matmul(tensor_A,tf.transpose(tensor_B))
mul
```

```python
#Multiply the two tensors you created in 3 using dot product.
mul_dot=tf.tensordot(tensor_A,tf.transpose(tensor_B),axes=1)
mul_dot

#Matmul and tensordot gives the same result in this case.
```

```python
#Create a tensor with random values between 0 and 1 with shape [224, 224, 3].
C=tf.random.uniform(minval=0,maxval=1,shape=[224,224,3])
C
```

```python
#Find the min and max values of the tensor you created in 6 along the first axis.
min=tf.reduce_min(C,axis=0)
min
```

For further understanding:For a given matrix,

[[3, 7, 1],
 [2, 8, 6],
 [5, 4, 9]]


Min values along axis 0: [2, 4, 1]
Max values along axis 0: [5, 8, 9]

```python
D=tf.constant([
 [3, 7, 1],
 [2, 8, 6],
 [5, 4, 9]])
tf.reduce_min(D,axis=0)
tf.reduce_min(D,axis=1)
```

```python
#Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].
F=tf.random.uniform(minval=2,maxval=200,shape=[1, 224, 224, 3])
print(F,F.shape)
F_squeezed=tf.squeeze(F)
print(F_squeezed,F_squeezed.shape)

#squeeze removes the dimensions of size 1 from the shape of a tensor.
```

```python
#Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.
E=tf.constant([4,6,2,3,1,7,8,10,45,0])
tf.argmax(E)
```

```python
#One-hot encode the tensor you created in 9.
E_onehot=tf.one_hot(E,depth=10)
E_onehot
```
